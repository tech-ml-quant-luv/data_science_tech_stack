### Handling data with classes 

```python
class MultiTickerHandler:
    
    def __init__(self, ticker ):
        self.ticker = ticker
        self.ltp_list = []

    def append_ltp(self,ltp):
        self.ltp_list.append(ltp)
        return self
```

```python
create_instances_of_tickers = {}
for ticker in tickers:
    mth = MultiTickerHandler(ticker)
    create_instances_of_tickers[ticker] = mth
create_instances_of_tickers
```

```python
for rowdata in data:
    symbol = rowdata["symbol"]
    ltp = rowdata["ltp"]
    create_instances_of_tickers[symbol].append_ltp(ltp)
```
